```python
import os
from google.colab import userdata
from huggingface_hub import whoami

try:
    # Obtenemos el token y limpiamos espacios en blanco accidentales
    raw_token = userdata.get('HF_TOKEN')
    HF_TOKEN = raw_token.strip() if raw_token else None

    if HF_TOKEN:
        # Establecemos la variable de entorno para todo el cuaderno
        os.environ["HF_TOKEN"] = HF_TOKEN

        # Verificación inmediata
        user = whoami(token=HF_TOKEN)
        print(f"✅ Token VÁLIDO. Conectado como: {user['name']}")
    else:
        print("❌ El secreto 'HF_TOKEN' está vacío.")

except Exception as e:
    print(f"❌ Error de validación: {e}")
    print("Revisa tu secreto 'HF_TOKEN' en el icono de la llave (🔑). Escribe el token nuevo y vuelve a ejecutar esta celda.")
```

```python
from huggingface_hub import InferenceClient

# Re-inicializando el cliente con el token correcto
client = InferenceClient(
    provider="hf-inference",
    api_key=os.environ["HFF_TOKEN"]
)
```

```python
SYSTEM_PROMPT = """
You are ChiquiQuest AI.

Transform career goals into RPG progression systems.

Generate:

1. Class
2. Level
3. XP
4. Skill Tree
5. Main Quests
6. Side Quests
7. Achievements
8. Roadmap

Format everything beautifully using markdown.

Return practical recommendations.
"""
```

```python
def generate_devquest(user_goal):
    # Reemplazado por el modelo de 7B para garantizar estabilidad en la API gratuita
    completion = client.chat.completions.create(
        model="Qwen/Qwen2.5-7B-Instruct",
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user", "content": user_goal}
        ],
        max_tokens=1200,
        temperature=0.8
    )
    return completion.choices[0].message.content
```

```python
def calculate_xp(text):

    xp = 0

    if "AWS" in text:
        xp += 300

    if "Docker" in text:
        xp += 200

    if "Kubernetes" in text:
        xp += 300

    if "Serverless" in text:
        xp += 250

    return xp
```

```python
def get_badges(text):
    badges = []
    text_lower = text.lower()

    # Cloud & Infrastructure
    if "aws" in text_lower:
        badges.append("☁️ Cloud Adventurer")
    if "s3" in text_lower or "storage" in text_lower:
        badges.append("📦 S3 Guardian")
    if "ec2" in text_lower or "compute" in text_lower:
        badges.append("⚡ EC2 Commander")

    # Development & Architecture
    if "full stack" in text_lower:
        badges.append("⚔️ Full Stack Paladin")
    if "frontend" in text_lower or "react" in text_lower:
        badges.append("⚛️ UI Visionary")
    if "backend" in text_lower or "node" in text_lower:
        badges.append("🚀 API Architect")

    # Advanced Skills
    if "docker" in text_lower or "kubernetes" in text_lower:
        badges.append("🐳 Orchestration Master")
    if "serverless" in text_lower or "lambda" in text_lower:
        badges.append("🔥 Serverless Wizard")

    return badges if badges else ["🎯 Quest Initiate"]
```

```python
def devquest(goal):
    result = generate_devquest(goal)
    xp = calculate_xp(goal)
    badges = get_badges(goal)

    # Mejoramos el formato de las estadísticas
    badges_str = "\n".join([f"* {b}" for b in badges])

    extra = f"""

---
# 📊 Panel de Personaje

### ✨ XP Acumulada: `{xp} pts`

### 🏆 Logros Desbloqueados
{badges_str}

---
*Sigue completando misiones para subir de nivel.*
"""
    return result + extra
```

```python
import gradio as gr

demo = gr.Interface(
    fn=devquest,
    inputs=gr.Textbox(
        lines=4,
        label="Career Goal"
    ),
    outputs=gr.Markdown(),
    title="🎮 ChiquiQuest AI",
    description="Transform your career into an RPG adventure"
)

demo.launch(share=True)
```
